```python
## create a simple txt file

import os
os.makedirs("../data/text_files", exist_ok=True)
```

```python
## Sample files
files = {
    "python.txt": """
Python is a popular programming language used for data science,
machine learning, web development, and automation.
""",
    "rag.txt": """
Retrieval-Augmented Generation (RAG) combines information retrieval
with large language models to provide accurate answers from documents.
""",
    "ai.txt": """
Artificial Intelligence (AI) enables machines to perform tasks
that typically require human intelligence, such as reasoning and learning.
"""
}

# Write files
for filename, content in files.items():
    with open(f"../data/text_files/{filename}", "w", encoding="utf-8") as f:
        f.write(content.strip())

print("Sample text files created successfully!")
```

```python
### TextLoader

### TextLoader

from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python.txt",encoding="utf-8")
document = loader.load()
print(document)
```

```python
from langchain_community.document_loaders import DirectoryLoader
dir_loader = DirectoryLoader(
    "../data/text_files",
    loader_cls=TextLoader,
    glob="*.txt",
    show_progress=False)

documents = dir_loader.load()
print(documents)
```

```python
## load all the text files from the directory

from langchain_community.document_loaders import PyPDFLoader,PyMuPDFLoader
#from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/pdf",
    loader_cls=PyPDFLoader,
    glob="**/*.pdf",
    show_progress=False
)

pdf_documents = dir_loader.load()
print(pdf_documents)
```
